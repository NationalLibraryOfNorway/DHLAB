```python
import dhlab.module_update as mu
mu.update('dhlab_v2')
import dhlab_v2 as d2
import pandas as pd
```

```python
corpus = d2.document_corpus(doctype = 'digibok')
```

```python
freqs = d2.frequencies(list(corpus.urn))
```

```python
f = pd.DataFrame(freqs)
```

```python
f
```

```python
words = f.sort_values(by = f.columns[0]).head(100000).index
len(words)
```

```python
%%time
d2.ngram_book(word = words, period = (1950, 1960)).transpose().sort_values(by = '1950', ascending = False)
```
